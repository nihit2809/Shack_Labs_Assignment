**Importing necessary libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Reading the data**

```python
df_amz=pd.read_csv("amz_com-ecommerce_sample.csv",encoding= 'unicode_escape')
df_flip=pd.read_csv("flipkart_com-ecommerce_sample.csv")
```

**Renaming the different columns as asked**

```python
df_amz.rename(columns={"product_name":"Product Name in Amazon", "retail_price":"Retail Price in Amazon","discounted_price":"Discounted Price in Amazon"},inplace=True)
```

```python
df_flip.rename(columns={"product_name":"Product Name in Flipkart", "retail_price":"Retail Price in Flipkart","discounted_price":"Discounted Price in Flipkart"},inplace=True)
```

**Merging the two dataframes with outer join so that we dont lose any products**

```python
df=pd.merge(df_amz,df_flip,how='outer').fillna("Not available")
```

```python
df.shape
```

**Dropping unecessary columns**

```python
df.drop(['uniq_id','crawl_timestamp','product_url','product_category_tree','is_FK_Advantage_product','pid','image','description','product_rating','overall_rating','brand','product_specifications'],axis=1,inplace=True)
```
